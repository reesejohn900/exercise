```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
train = pd.read_csv('/content/drive/My Drive/data/titanic/train.csv')
train.head()
```

```
train['Age'].fillna(24.0,inplace=True)
```

```
train['Cabin'].fillna('B98',inplace=True)
```

```
train['Embarked'].fillna('S',inplace=True)
```

```
train.isna().sum()
```

```
train_1 = train.drop(['PassengerId','Name','Ticket','Survived'],axis=1)
```

```
x_train = pd.get_dummies(train_1[['Sex','Cabin','Embarked']])
x_train
```

```
x_train['Pclass'] = train['Pclass']
x_train['Age'] = train['Age']
x_train['SibSp'] = train['SibSp']
x_train['Parch'] = train['Parch']
x_train['Fare'] = train['Fare']
```

```
y_train = train['Survived']
```

```
test = pd.read_csv('/content/drive/My Drive/data/titanic/test.csv')
test
```

```
test['Age'].fillna(test['Age'].mean(),inplace=True)
```

```
test['Cabin'].fillna('B98',inplace=True)
```

```
test['Fare'].fillna(test['Fare'].mean(),inplace=True)
```

```
test['Embarked'].fillna('S',inplace=True)
```

```
test.isna().sum()
```

```
x_test = pd.get_dummies(test[['Sex','Cabin','Embarked']])

x_test['Pclass'] = test['Pclass']
x_test['Age'] = test['Age']
x_test['SibSp'] = test['SibSp']
x_test['Parch'] = test['Parch']
x_test['Fare'] = test['Fare']
```

```
missing_cols = set(x_train.columns)-set(x_test.columns)
```

```
for c in missing_cols:
    x_test[c] = 0
```

```
missing_cols_1 = set(x_test)-set(x_train)
```

```
for c in missing_cols_1:
    x_train[c]=0
```

```
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

```
model.fit(x_train,y_train)
```

```
y_preds = model.predict(x_test)
```

```
grid = {'n_estimators':[100,200,600],
        'max_depth':[None],
        'max_features':['auto','sqrt'],
        'min_samples_split':[6],
        'min_samples_leaf':[1,2]}
```

```
from sklearn.model_selection import GridSearchCV

gs_model = GridSearchCV(estimator=model,
                        param_grid=grid,
                        cv=5,verbose=2)
```

```
gs_model.fit(x_train,y_train)
```

```
gs_model.best_params_
```

```
model = RandomForestClassifier(max_depth=None,
 max_features='sqrt',
 min_samples_leaf= 1,
 min_samples_split= 6,
 n_estimators= 100)
```

```
np.random.seed(42)
model.fit(x_train,y_train)
```

```
np.random.seed(42)

y_preds = model.predict(x_test)
y_preds
```
