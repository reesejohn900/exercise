```python
heart_disease = pd.read_csv('13.1 heart-disease.csv')
heart_disease.head()
```

```python
heart_disease['age'].plot(kind='hist',bins=50);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
plt.style.use('seaborn-whitegrid')
fig,ax = plt.subplots(figsize=(10,5))
Heart_plot = ax.scatter(over_50['age'],over_50['chol'],c=over_50['target'],cmap='winter')
ax.set(title = 'Heart Disease And Cholestrol Level',
       ylabel='Cholestrol',
       xlabel='Age')
ax.legend(*Heart_plot.legend_elements(),title='Target')
ax.axhline(over_50['chol'].mean(),linestyle='--',c='red')
ax.set_xlim([50,80]);
```
